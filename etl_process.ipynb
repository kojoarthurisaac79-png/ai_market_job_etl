```python
import re

data['job_title'] = data['job_title'].str.replace(r'[^A-Za-z0-9\s]', '', regex=True)
```

```python
import re
# Extract lower and upper bounds
data['salary_min'] = data['salary_range_usd'].str.extract(r'(\d{2,3}[,\d]*)').replace(',', '', regex=True).astype(float)
data['salary_max'] = data['salary_range_usd'].str.extract(r'-(\d{2,3}[,\d]*)').replace(',', '', regex=True).astype(float)
```

```python
import re
data['experience_level_clean'] = data['experience_level'].str.extract(r'(Junior|Mid|Senior|Entry|Lead)', expand=False)
```

```python
import re
# Remove extra spaces around skills
data['skills_required'] = data['skills_required'].str.replace(r'\s*,\s*', ',', regex=True)
```

```python
import re
data['location'] = data['location'].str.replace(r'[^A-Za-z0-9,\s]', '', regex=True)
```

```python
from sqlalchemy import create_engine
import pandas as pd
import mysql.connector

file_path = "ai_job_market.csv"
df = pd.read_csv(file_path)

# Ensure database exists
conn = mysql.connector.connect(
    host="localhost",
    port=3306,
    user="root",
    password="1234"
)
cursor = conn.cursor()
cursor.execute("CREATE DATABASE IF NOT EXISTS ai_job_db;")
conn.commit()
cursor.close()
conn.close()

# Connect using SQLAlchemy
engine = create_engine("mysql+mysqlconnector://root:1234@localhost:3306/ai_job_db")

# Load DataFrame
df.to_sql("ai_job_db", con=engine, if_exists="replace", index=False)
print(" Successfully loaded data into 'ai_job_db' table!")
```
